```python
import pandas as pd

dates = [
  '03/01/2021',
  '17/02/2020',
  '25/03/2019',
  '08/04/2022',
  '14/05/2018',
  '29/06/2023',
  '10/07/2020',
  '22/08/2021',
  '05/09/2017',
  '31/10/2024'
]

df = pd.DataFrame({'date': dates})
```

```python
# “Data Scientists spend 80% of the time cleaning data and only 20% modeling.”
# normalization
# categorical encoding
# feature engineering
```

```python
df['month'] = df['date'].str[3:5]
```
